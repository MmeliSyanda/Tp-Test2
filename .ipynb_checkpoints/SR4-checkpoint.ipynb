```python
import pandas as pd
myDataset = pd.read_csv('C:/Users/Laptop C/Downloads/Informal_Settlement_Programme_-1898186875815614491.csv')

missing_data = myDataset.isnull().sum()


myDataset['ALT_NAME'] = myDataset['ALT_NAME'].fillna('Unknown')
missing_data2 = myDataset.isnull().sum()


def insertion_sort(arr):
    """
    Function to perform insertion sort on an array.

    Parameters:
    - arr: The input array to be sorted.
    """
    # Traverse through the array starting from the second element
    for i in range(1, len(arr)):
        # Current element to be inserted at the correct position
        key = arr[i]
        # Initialize the index to compare with
        j = i - 1
        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        # Insert the key into its correct position
        arr[j + 1] = key

# Load the dataset and select the column to be sorted
column_to_sort = myDataset['EST_HHC'].tolist()  # Convert the column to a list

# Perform insertion sort on the selected column (in ascending order)
insertion_sort(column_to_sort)

# Display the sorted column
print("Sorted column (ascending order):", column_to_sort)

def linear_search(arr, target):
    """
    Function to perform linear search in an unsorted array.

    Parameters:
    - arr: The input array to be searched.
    - target: The target element to search for.

    Returns:
    - index: The index of the target element if found, else -1.
    """
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    """
    Function to perform binary search in a sorted array.

    Parameters:
    - arr: The input array to be searched (must be sorted).
    - target: The target element to search for.

    Returns:
    - index: The index of the target element if found, else -1.
    """
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Example usage:
# Assuming 'EST_HHC' is the column to be searched
column_to_search = myDataset['EST_HHC'].tolist()  # Convert the column to a list
target_element = 123  # Specify the target element to search for

# Perform linear search
linear_result = linear_search(column_to_search, target_element)
if linear_result != -1:
    print("Linear Search: Element found at index", linear_result)
else:
    print("Linear Search: Element not found")

# Perform binary search (assuming the dataset column is sorted)
column_to_search.sort()  # Sort the column for binary search
binary_result = binary_search(column_to_search, target_element)
if binary_result != -1:
    print("Binary Search: Element found at index", binary_result)
else:
    print("Binary Search: Element not found")
```
